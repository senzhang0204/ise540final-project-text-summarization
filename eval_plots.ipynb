### Model Comparison

```python
avg_simple.reset_index(inplace=True)
avg_simple
```

```python
avg_simple.iloc[6,0] = '7. Our_Model'
avg_simple
```

```python
avg_simple1 = pd.melt(avg_simple, id_vars = 'Summary_type')
avg_simple1.columns = ['Models', 'Metrics', 'value']
avg_simple1.head()
```

```python
sns.set(font_scale = 1.6)
sns.set_style("whitegrid")
sns.catplot(x = 'Models', y='value', 
            hue = 'Metrics',data=avg_simple1, 
            kind='bar', height=10, aspect=18/10)
# plt.xlabel('Models')
plt.ylabel('Metric Values')
plt.show()
```

```python
avg = eval.groupby("Summary_type").median() 
fle = pd.DataFrame(avg.f_score)
fle.reset_index(inplace=True)
fle.iloc[0,0] = '0. Ground_Truth'
fle.iloc[7,0] = '7. Our_Model'
fle.columns = ['Models', 'Flesch Score']
#fle.to_html('flesch.html')
fle.to_csv('flesch.csv')
fle
```

```python
# Flesch Score Comparison
plt.figure(figsize=(15,8))
plt.bar(fle['Models'], fle['Flesch Score'])
plt.xticks(rotation = 45, fontsize = 15)
plt.ylim((-600, 200))
plt.title('Model Flesch Score', fontsize=15)
plt.show()
```

```python
flesch_ease = pd.DataFrame(eval.groupby("Summary_type")['f_ease', 'f_grade_levels'].agg(lambda x:x.value_counts().index[0]))
flesch_ease.reset_index(inplace=True)
flesch_ease.columns = ['Models', 'Flesch_Ease', 'Flesch_Grade_Level']
#flesch_ease.to_html('flesch_ease.html')
flesch_ease.to_csv('flesch_ease.csv')
flesch_ease
```
